```python
import numpy as np
import pandas as pd
import requests
from pprint import pprint
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Polygon
```

```python
def get_geo_info(url):
    """ 
      La función recibe la url y retorna un geopandas.DataFrame
    """
    response = requests.get(url,)
    # Verificar el código de respuesta
    if response.status_code == 200:
        json_response, geometries = response.json(), []    
        for feature in json_response['features']:
            rings = feature['geometry']['rings']
            polygon = Polygon(rings[0])
            geometries.append(polygon)
        ##
        attributes = [feature['attributes'] for feature in json_response['features']]
        df = pd.DataFrame(attributes)
        gdf = gpd.GeoDataFrame(df, geometry=geometries)
        
        return gdf    
    else:
        print(f'ERROR IN RESPONSE: {response.status_code}')
##
url = "https://datosabiertos.bogota.gov.co/dataset/808582fc-ffc8-4649-8428-7e1fd8d3820c/resource/a5c8c591-0708-420f-8eb7-9f3147e21c40/download/unidadplaneamientolocal.json"
get_geo_info(url)
```

```python
### PILLE EL EJEMPLO DE chatgpt ###
url = "https://datosabiertos.bogota.gov.co/dataset/808582fc-ffc8-4649-8428-7e1fd8d3820c/resource/a5c8c591-0708-420f-8eb7-9f3147e21c40/download/unidadplaneamientolocal.json"
response = requests.get(url,)
json_response = response.json()
# Definir una lista vacía para almacenar las geometrías
geometries = []
# Recorrer las features del JSON y extraer las geometrías
for feature in json_response['features']:
    rings = feature['geometry']['rings']
    # Crear un objeto de tipo Polygon a partir de los anillos
    polygon = Polygon(rings[0])
    geometries.append(polygon)
print(geometries)
# Crear un DataFrame con los atributos
attributes = [feature['attributes'] for feature in json_response['features']]
df = pd.DataFrame(attributes) ## pandas.DataFrame
display(df.head())
# Crear un GeoDataFrame a partir del DataFrame y las geometrías
gdf = gpd.GeoDataFrame(df, geometry=geometries)
display(gdf.head())
# Visualizar el GeoDataFrame
fig, ax = plt.subplots(figsize=(60, 8), dpi=100)
# Plotear el GeoDataFrame con los parámetros de tamaño personalizados
gdf.plot(ax=ax)
# Mostrar el plot
plt.show()
```
